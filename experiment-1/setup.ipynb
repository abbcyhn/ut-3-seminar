## Results

```python
import pandas as pd
import matplotlib.pyplot as plt

tests = ['Reading in the file', 'Compute metrics of a column (MEAN)', 'Compute metrics of a column (STD)', 'Finding the unique value of a column', 'Cumulative sum of a column', 'Groupby Aggregation']
results = { 'Pandas'        : [106,     5.28,   67.8,   77.8,     65.4,   122.23],
            'Dask'          : [17,      40.38,  259.8,  29.5,     247.2,  182.2],
            'Modin (Ray)'   : [225,     31.56,  345.75, 167.3,    524.78, 993.6],
            'Vaex'          : [163,     343.2,  335.4,  524.1,    408.09, 2613.6]
        } 

results_df = pd.DataFrame(data=results, index=tests)

for test in results_df.index:
    plt.xlabel('Frameworks')
    plt.ylabel('Execution Time (seconds)')
    results_df.loc[test].plot(kind='bar', title=test)
    plt.show()
```
